```python
all_ranks = []
all_titles = []
all_urls = []
all_scores = []
```

```python
def get_page(driver):
    page=[]    
       
    movie_page = driver.find_elements('xpath',"//*[contains(@id, '#area')]")
    movie_names = [element.text for index, element in enumerate(movie_page) if index % 2 != 0]
    
    movie_url = [element.get_attribute('href') for index, element in enumerate(movie_page) if index % 2 != 0]
    
    movie_rank_page = driver.find_elements('xpath',"//*[contains(@class,'lightLink top-anime-rank-text rank')]")
    movie_ranks = [element.text for index, element in enumerate(movie_rank_page)]
    
    movie_score_page = driver.find_elements('xpath',"//*[contains(@class,'text on score-label score')]")
    movie_scores = [element.text for index, element in enumerate(movie_score_page)]
    
    return movie_names,movie_url,movie_ranks,movie_scores
```

```python
#dataframe
url = 'https://myanimelist.net/topanime.php'  # Replace with the actual URL

# Loop through pages
while True:
    # Navigate to the page
    driver.get(url)

    # Scrape the data from the page
    titles, urls, ranks, scores = get_page(driver)

    # Append the scraped data to the lists
    all_ranks.extend(ranks)
    all_titles.extend(titles)
    all_urls.extend(urls)
    all_scores.extend(scores)

    try:
        # Find the "Next" button and get the URL of the next page
        next_button = driver.find_element('xpath','//*[@class="link-blue-box next"]')
        url = next_button.get_attribute('href')
    except:
        # If there's no "Next" button, break the loop
        break

# Create a DataFrame
df1 = pd.DataFrame({
    'Rank': all_ranks,
    'Title': all_titles,
    'URL': all_urls,
    'Score': all_scores
})

# Print the DataFrame
df1
```

```python

def scrape_anime_page(url, max_retries=9):
    timeout = random.uniform(1.5, 2)  # Generate a random timeout between 1.5 and 2 seconds
    retries = 0
    
    while retries < max_retries:
        try:
            response = requests.get(url, timeout=timeout)
            response.raise_for_status()  # Raise HTTPError for bad responses
            break  # Exit loop if request is successful
        except requests.exceptions.RequestException as e:
            retries += 1
            print(f"Request failed for {url} with timeout {timeout}: {e}. Retrying ({retries}/{max_retries})...")
            time.sleep(2 ** retries)  # Exponential backoff
    
    if retries == max_retries:
        print(f"Failed to retrieve {url} after {max_retries} attempts.")
        return None
    
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Print the HTML content for debugging
    # print(soup.prettify())
    time.sleep(random.uniform(0.6, 1.2))
    
    # Extract the desired information using more specific selectors
    title_element = soup.find('h1', class_='title-name h1_bold_none')
    anime_title = title_element.text.strip() if title_element else 'Title not found'

    genre_elements = soup.find_all('span', itemprop='genre')
    anime_genres = [genre.text.strip() for genre in genre_elements] if genre_elements else 'Genres not found'

    # New fields to extract
    episodes_element = soup.find('span', string='Episodes:')
    anime_episodes = episodes_element.find_next_sibling(string=True).strip() if episodes_element else 'Episodes not found'

    premiered_element = soup.find('span', string='Premiered:')
    if premiered_element and premiered_element.find_next_sibling('a'): 
        premiered_sibling = premiered_element.find_next_sibling('a') 
        anime_premiered = premiered_sibling.text.strip() if premiered_sibling and isinstance(premiered_sibling, str) else 'Premiered season not found' 
    else: anime_premiered = 'Premiered season not found'
        
    studio_element = soup.find('span', string='Studios:')
    anime_studio = studio_element.find_next_sibling('a').text.strip() if studio_element else 'Studio not found'

    type_element = soup.find('span', string='Type:')
    anime_type_ = type_element.find_next_sibling('a').string.strip() if type_element else 'Type not found' 
    
    source_element = soup.find('span', string='Source:')
    anime_source = source_element.find_next_sibling('a').text.strip() if source_element and source_element.find_next_sibling('a') else 'Source not found'
    
    return {
        'Title': anime_title,
        'Genres': ', '.join(anime_genres),
        'Episodes': anime_episodes,
        'Premiered': anime_premiered,
        'Studio': anime_studio,
        'Type': anime_type_,
        'Source': anime_source,
    }

# Initialize list to hold scraped data
all_data = []

# Loop through URLs and scrape data
for url in df['URL'][9659:10000]:
    data = scrape_anime_page(url)
    if data:
        all_data.append(data)
        df_current = pd.DataFrame(all_data)
        df_current.to_csv('anime_data_17.csv')
        print(f"{url} saved")
    else:
        print(f"{url} not saved due to repeated failures")

    # Random delay to avoid CAPTCHA
    delay = random.uniform(0.5, 0.6)
    print(f"Delaying for {delay} seconds to avoid CAPTCHA...")
    time.sleep(delay)
```
